```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt;plt.style.use('fivethirtyeight')
import seaborn as sns;
```

```python
df = pd.read_csv('Runners_all_results.csv', error_bad_lines=False, encoding='cp1252')
df
```

```python
df.info()
```

```python
df.columns
```

```python

f = plt.figure(figsize = (14, 10))
sns.distplot(df['AgeOnRaceDay'], bins = 60)

```

```python
plt.figure(figsize = (20, 10))
sns.countplot('AgeOnRaceDay', data = df, hue = 'Gender')
```

```python
p = df.groupby('CountryOfResName').count()['City'].sort_values(ascending=False).head(20)[1:]
q = sns.barplot(p.index, p, palette='rainbow')
q.figure.set_size_inches(16,8)
q.set_title("Most popular Country")
```

```python
l = sns.boxplot(df['Gender'], df['AgeOnRaceDay'], palette="coolwarm")
l.set_title("Distribution of Official Time and Gender")
```

```python
df['OfficialTime'] = pd.to_timedelta(df['OfficialTime'])
```

```python
k = plt.figure(figsize = (20, 8))
k = sns.boxplot(x='AgeOnRaceDay', y=df['OfficialTime'].apply(lambda x: x.total_seconds()/3600), data=df, palette="Blues")
k.set_title("Distribution of the Official Time and Age")
```
