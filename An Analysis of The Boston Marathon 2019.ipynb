```python
x = f[['AgeOnRaceDay']]
y = f['OfficialTime (Hours)']

from sklearn import linear_model
lm = linear_model.LinearRegression()
lm.fit(x, y)
lm.score(x, y)

from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y, lm.predict(x))
print(f"r_squared: {round(r2, 3)}")
#print("Mean Squared Error:", mean_squared_error(y, lm.predict(x)))
```

```python
plt.figure(figsize = (15, 8))
plt.scatter(x = 'AgeOnRaceDay',
             y = 'OfficialTime (Hours)',
             data = f,
           s = 100)
plt.plot(x, lm.predict(x), c= 'm')

plt.title('Age and Finish Time (Hours)', size = 25)
plt.xlabel("Age", size = 25)
plt.ylabel('Finish time (Hours)', size = 25)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.legend(['Linear Regression', 'Data'], fontsize = 20)
```

```python
from sklearn.preprocessing import PolynomialFeatures
x = f[['AgeOnRaceDay']]
y = f['OfficialTime (Hours)']

polynomial_features= PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x)
model = linear_model.LinearRegression()
model.fit(x_poly, y)

r2 = r2_score(y, model.predict(x_poly))
print(f"r_squared: {round(r2, 3)}")
```

```python
plt.figure(figsize = (15, 8))
plt.scatter(x = 'AgeOnRaceDay',
             y = 'OfficialTime (Hours)',
             data = f,
           s = 100)
plt.plot(x, model.predict(x_poly), c = 'm')

plt.title('Age and Finish Time (Hours)', size = 25)
plt.xlabel("Age", size = 25)
plt.ylabel('Finish time (Hours)', size = 25)
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.legend(['Quadratic Regression', 'Data'], fontsize = 20)
```

```python
from scipy import optimize

def linear_function(x, a, b):
    return a * x + b

def quadratic_function(x, a, b, c):
    return a * (x**2) + b * x +c

X, y = list(f['AgeOnRaceDay']), f['OfficialTime (Hours)'].tolist()
params, _ = optimize.curve_fit(linear_function, X, y)
print('Estimated function: {0:.3f} * x + {1:.3f}'.format(*params))

params_2, _ = optimize.curve_fit(quadratic_function, X, y)
print('Estimated function: {0:.3f} * x^2 + {1:.3f} * x + {2:0.3f}'.format(*params_2))
```

```python
plt.figure(figsize = (16, 8))
plt.scatter(f['AgeOnRaceDay'], f['OfficialTime (Hours)'])
estimate = [linear_function(x, *params) for x in X]
plt.plot(f['AgeOnRaceDay'], estimate, color='red', label='Model')
estimate_2 = [quadratic_function(x, *params_2) for x in X]
plt.plot(f['AgeOnRaceDay'], estimate_2, color='m', label='Model')
plt.xticks(rotation = 90)
plt.legend(['Linear Regression', 'Quadratic Regression', 'Official Time'])
```

```python
male = df[df['Gender'] == 'M']
female = df[df['Gender'] == 'F']
```

```python
n1 = len(male['OfficialTime (Hours)'])
n2 = len(female['OfficialTime (Hours)'])
print('Number of male n1 =', n1)
print('Number of female n2 =', n2)

mean_male = male['OfficialTime (Hours)'].mean().round(3)
mean_female = female['OfficialTime (Hours)'].mean().round(3)
print(f"Male Official time sample mean {mean_male}")
print(f"Feale Official time sample mean {mean_female}")

std_male = male['OfficialTime (Hours)'].std().round(3)
std_female = female['OfficialTime (Hours)'].std().round(3)
print(f"Male Official time std {std_male}")
print(f"Female Official time std {std_female}")

```

Doing a z test since two samples are large enough



```python
import math
z_score = (mean_male - mean_female)/math.sqrt(std_male**2/n1+std_female**2/n2)
print('Z value:', z_score)
```

The P-Value is < .00001.

The result is significant at p < .05.

So male are quicker than female
